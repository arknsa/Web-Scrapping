```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://steamcharts.com/top'
page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
print(soup)
```

```python
soup.find('table')
```

```python
soup.find_all('table')[0]
```

```python
soup.find('table', class_ = 'common-table')
```

```python
table = soup.find_all('table')[0]
table
```

```python
top_games = table.find_all('th')
top_games
```

```python
top_games_titles = [title.text.strip() for title in top_games]

print(top_games_titles)
```

```python
df = pd.DataFrame(columns = top_games_titles)

df
```

```python
column_data = table.find_all('tr')
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
    
df
```

```python
columns_drop = ['', 'Last 30 Days']
df.drop(columns=columns_drop, axis=1)
```
